## Correcção de programas -- Pré e Pós-condições

### Exercício:

Dadas as seguintes definições de funções, indique as pré e pós-condições de cada uma:

```ocaml
let rec fact x =
  if x = 0 then
    1
  else x * fact (x - 1)

let rec gcd u v =
  if v = 0 then
    u
  else gcd v (u mod v)

  let division a b =
    let rec division' q r =
      if r < b then
        (q, r)
      else
        division' (q + 1) (r - b) a b
    in
    division' 0 a 
```
